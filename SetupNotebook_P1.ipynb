### Estimate Image statistics

```python
import numpy as np
```

```python
# prompt: Generate a function that will estimate image statistics: mean, max min, standard deviation of pixels from tumor and no tumor images

import torch

def estimate_image_statistics(image,label):
    """
    Estimates the mean, max, min, and standard deviation of pixel values from tumor and no tumor images.

    Args:
        train_dataset: A PyTorch ImageFolder object containing the MRI dataset.

    Returns:
        A dictionary containing the estimated statistics for tumor and no tumor images.
    """
    image = torch.tensor(np.array(image)).float()
    # Initialize variables
    image_mean = torch.zeros(3)
    image_max = torch.zeros(3)
    image_min = torch.ones(3) * 255
    image_std = torch.zeros(3)




    # Calculate statistics for image


    image_mean = torch.mean(image)
    image_max = torch.max(image)
    image_min = torch.min(image)


    image_std = torch.std(image)

    if label == 1:
      nature = "tumor"
    else:
      nature = "notumor"
    print(label)



    # Return the estimated statistics
    return [
        nature, {"mean":image_mean,
            "max": image_max,
            "min": image_min,
            "std": image_std}]

```

### Histogram for MRI intensity

```python
# prompt: a function that can plot histogram for pixel intensities in an image

import matplotlib.pyplot as plt

def plot_histogram(image):
  """
  Plots a histogram of pixel intensities in an image.

  Args:
    image: A NumPy array representing the image.
  """
  # Flatten the image array
  image = torch.tensor(np.array(image)).float()
  flat_image = image.flatten()

  # Calculate the histogram
  hist, bins = np.histogram(flat_image, bins=256)

  # Plot the histogram
  plt.hist(flat_image, bins=bins)
  plt.xlabel("Pixel Intensity")
  plt.ylabel("Frequency")
  plt.title("Histogram of Pixel Intensities")
  plt.show()
```

```python
#  code for image equalization with cv2

import matplotlib.pyplot as plt
import numpy as np
import cv2

def equalize_image(image,label):
  """
  Performs histogram equalization on an image.

  Args:
    image: A NumPy array representing the image.

  Returns:
    The equalized image as a NumPy array.
  """
  tumor = "tumor" if label == 0 else "notumor"
  title = "Equalized "+ tumor + " image "
  # Convert the image to grayscale
  gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  # Perform histogram equalization
  equ_image = cv2.equalizeHist(gray_image)

  # Convert the image back to color
  equ_color_image = cv2.cvtColor(equ_image, cv2.COLOR_GRAY2BGR)

  return equ_color_image,title

```

```python
# Visualize the images
titles = []
fig, axes = plt.subplots(1, 2, figsize=(7, 2))
equ_image,title = equalize_image(np.array(train_dataset[1601][0]),train_dataset[1601][1])
titles.append( title.split(" ")[1])
titles.append(title)
for i, image in enumerate([np.array(train_dataset[1601][0]),equ_image]):
    ax = axes[i]
    ax.imshow(image)
    ax.set_title(titles[i])
    ax.axis("off")

plt.show()
```

### Phantom images

```python
# A shepp_logan_phantom function for phantom image

import matplotlib.pyplot as plt
import torch
from phantominator import shepp_logan

# Generate a Shepp-Logan phantom with default parameters
phantom = shepp_logan(N=256)

# Print the phantom shape
print(phantom.shape)

# Display the phantom using matplotlib
plt.imshow(phantom, cmap="gray")
plt.axis("off")
plt.show()

```

```python
# Perform a 2D Fourier transform on the phantom
fft_phantom = np.fft.fft2(phantom)

# Shift the zero frequency component to the center of the spectrum
fft_phantom_shifted = np.fft.fftshift(fft_phantom)


```

```python
# Apply blur in the frequency space using convolutions.

# Define the size of the Gaussian kernel
kernel_size = 256

# Create a Gaussian kernel
kernel = np.outer(
    np.exp(-np.linspace(-1, 1, kernel_size)**2),
    np.exp(-np.linspace(-1, 1, kernel_size)**2)
)

# Normalize the kernel
kernel /= np.sum(kernel)

# Apply the blur in the frequency space by convolving the kernel with the shifted FFT of the phantom
blurred_fft_phantom = fft_phantom_shifted*kernel


# Display the blurred phantom
plt.imshow(np.abs(blurred_fft_phantom), cmap="gray")
plt.axis("off")
plt.show()
```

```python
# Shift the zero frequency component back to the origin
blurred_fft_phantom_unshifted = np.fft.ifftshift(blurred_fft_phantom)

# Perform an inverse 2D Fourier transform to obtain the blurred phantom
blurred_phantom = np.fft.ifft2(blurred_fft_phantom_unshifted)

# Display the blurred phantom
plt.imshow(np.abs(blurred_phantom), cmap="gray")
plt.axis("off")
plt.show()
```
